**Cleaning DataFrame Busquedas**

* Columns selection 
* Create column trip & busquedas *(Una linea del archivo es una busqueda, la usaremos luego para contar el numero de busquedas cuando agrupemos)*
* Eliminate columns not usable

```python
busquedas.head()
```

```python
busquedas = pd.read_csv('searches.csv.bz2',sep="^",nrows=100000, usecols=('Origin','Destination'))
```

```python
busquedas['trip']=busquedas['Origin']+str('-')+busquedas['Destination']
busquedas['busquedas']=1
busquedas.drop(['Origin'], axis = 1, inplace = True)
busquedas.drop(['Destination'], axis = 1, inplace = True)
```

```python
busquedas.head(100)
```

**Cleaning DataFrame Reservas**

* Columns selection 
* Eliminate cancellations
* Correct number of characters in dep_port & arr_ports
* Create column trip
* Eliminate columns not usable

```python
reservas.head()
```

```python
reservas.columns
```

```python
reservas= pd.read_csv ('bookings.csv.bz2', nrows= 100000, sep='^',usecols = ('pax','dep_port','arr_port'))
```

```python
# elimino las cancelaciones (pax ==0)

reservas=reservas[reservas['pax']==1]
```

```python
# corrijo las columnas de texto para tener solo 3 caracteres

reservas['arr_port']=reservas['arr_port'].str.strip().str.upper()
reservas['dep_port']=reservas['dep_port'].str.strip().str.upper()
```

```python
reservas['trip']=reservas['dep_port']+str('-')+reservas['arr_port']
```

```python
reservas.drop(['dep_port'], axis = 1, inplace = True)
reservas.drop(['arr_port'], axis = 1, inplace = True)
```

```python
reservas.head()
```

**Matching Searching & Bookings**
* Agrupo por trip en ambas tablas

```python
busquedas_agrupadas = busquedas.groupby('trip').count()
busquedas_agrupadas.head()
```

```python
reservas_agrupadas = reservas.groupby('trip').count()
reservas_agrupadas.head()
```

```python
resultado = pd.merge(busquedas_agrupadas,reservas_agrupadas, left_index=True, right_index=True)
resultado
```
